#### Explorando uma base de dados
Vamos explorar a base de dados `melb_data.csv`, da pasta `data` desse diretório. Essa base contém dados sobre preços de casas. Vamos usar o `DataFrame`, que armazena um tipo de dado semelhante a uma tabela.

```python
# salvando o caminho para o arquivo em uma variávell
melbourne_file_path = './data/melb_data.csv'
# lendo e armazenando os dados em um DataFrame chamado melbourne_data
melbourne_data = pd.read_csv(melbourne_file_path) 
# printa um summário dos dados em melbourne_data
melbourne_data.describe()
```

##### Explicação dos valores retornados
Esse resultado retorna 8 valores para cada coluna do conjunto de dados original.
* count: quantas linhas do dataset possuem valores não zerados
* mean: média dos valores daquela coluna
* std: desvio padrão dos valores daquela coluna (Reveja esse conceito depois)
* min: Valor mínimo
* 25%: Valor do 25º percentil (ou 1º quartil)
* 50%: Valor do 50º percentil (ou 2º quartil)
* 75%: Valor do 75º percentil (ou 3º quartil)
* max: Valor máximo
